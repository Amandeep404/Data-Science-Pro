```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8-pastel')

import warnings
warnings.filterwarnings('ignore')
```

```python
# we drop the rows where rating column is null
df.dropna(subset=['rating_count'], inplace=True)
```

```python
mean_rating = df[df['rating'] !='|']['rating'].astype('float').mean()
df['rating'] = df['rating'].replace('|', mean_rating).astype('float')
```

```python
df['discount_percentage'] = df['discount_percentage'].str.removesuffix('%')
```

```python
df['rating_count'] = df['rating_count'].str.replace(',', '')
```

#### Converting Data type to int and float of all the required columns

```python
df.discounted_price =  df.discounted_price.str.replace('‚Çπ', '')
df.discounted_price =  df.discounted_price.str.replace(',', '')
df['discounted_price'] = df.discounted_price.astype('float')
```

```python
df.actual_price = df.actual_price.str.replace('[‚Çπ,]', '',regex=True)
df.actual_price = df.actual_price.astype('float')
```

```python
df.discount_percentage =  df.discount_percentage.astype(int)
```

```python
df.rating_count = df.rating_count.astype(int)
```

```python
df.info()
```

### 1. What is the average rating for each product category ?

The average rating of all the product categories are given below. The hihest rating is given for Office Products category

```python
df['main_category'] = df['category'].str.split('|').str.get(0)
avg_rating = df.groupby('main_category')['rating'].mean().round(2)
colors = ['royalblue', 'palegreen', 'lightcoral', 'orchid', 'wheat'] 
plt.bar(x=avg_rating.index, height= avg_rating, color = colors)
plt.title('Average Rating by Categories', color = 'red')
plt.xticks(rotation=90)
for i, v in enumerate(avg_rating):
    plt.text(i, v, str(v), ha='center', va='bottom')

plt.show()
```

### 2. What are the top rating_count products by category ?
The top rating products in each category is shown in the visualization below. The product with highest rating count is Amazon Basics Flexible premium


```python
top_product_ratings = df.groupby('main_category').agg({'product_name': 'first', 'rating_count': 'max'}).reset_index()

plt.figure(figsize=(10, 6))
barplot = sns.barplot(x='rating_count', y=top_product_ratings['product_name'].str.split().str[:3].str.join(' '), 
            hue='main_category', data=top_product_ratings)

for index, row in top_product_ratings.iterrows():
    barplot.text(row['rating_count'], index, str(row['rating_count']), color='black', va='center')

plt.xlabel('Rating Count')
plt.ylabel('Product Name')
plt.title('Top Rated Products by Category')
plt.legend(title='Main Category', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

### 3. What is the distribution of discounted prices vs. actual prices? 

```python
plt.subplot(1,2,1)
sns.kdeplot(data = df[df['discounted_price']<7000]['discounted_price'], label = 'Discounted price', shade = True, color='skyblue')
sns.kdeplot(data = df[df['actual_price']<7000]['actual_price'], label = 'Actual price', shade = True, color= 'orange')
plt.title('Distribution of Discounted price v/s Actual price', color = 'red')
plt.legend()


plt.subplot(1,2,2)
sns.histplot(data = df[df['discounted_price']<7000]['discounted_price'], label = 'Discounted price', kde=True, color='skyblue')
sns.histplot(data = df[df['actual_price']<7000]['actual_price'], label = 'Actual price', kde=True, color= 'orange')
plt.legend()

plt.tight_layout()
plt.show()

```

### 4. How does the average discount percentage vary across categories?

```python
d_price = df.groupby('main_category')['discounted_price'].mean().round(2)
a_price = df.groupby('main_category')['actual_price'].mean().round(2)
avg_discount_percentage = ((1 - d_price/a_price )* 100).round(2)

barplot = sns.barplot(x = avg_discount_percentage.values, y = avg_discount_percentage.index)
plt.title('Discount Percentage in Categories')
plt.ylabel('Categories')
plt.xlabel('Discount (%)')

for index,value in enumerate(avg_discount_percentage.values):
    barplot.text(x = value + 0.5 , y = index + 0.1, s = f'{value}%', ha = 'left')
plt.show()
```

### 8.  What is the correlation between discounted_price and rating?
The correlation between discouneted price and rating is 0.12. We see a trend that products with higher discounted price have higher rating.


```python
df['discount_percentage'] = ((1 - (df['discounted_price'] / df['actual_price'])) * 100).round(1)
```

```python
corr_discounted_price_and_rating = df['discounted_price'].corr(df['rating'])
sns.lineplot(x = df['rating'], y= df['discounted_price'], color='lightcoral')
plt.title('Correlation between Discounted Price and Rating')
plt.xlabel('Rating')
plt.ylabel('Discounted Price')
plt.show()
print(f'The correlation between discounted price and rating is  {corr_discounted_price_and_rating.__round__(2)}')
```

### 9. What are the Top 5 categories based on the highest ratings ?


```python
highest_rating_cat = df.groupby('main_category')['rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(12,5))
barplot = sns.barplot(x = highest_rating_cat.values, y = highest_rating_cat.index)
for index, value in enumerate(highest_rating_cat.values):
    barplot.text(x = value + 0.03, y = index, s = f'{round(value,1)}')
plt.title('TOP 5 Categories')
plt.xlabel('Rating')
plt.ylabel('Categories')
plt.show()
```
